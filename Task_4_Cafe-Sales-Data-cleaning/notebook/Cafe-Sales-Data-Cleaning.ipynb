```python
df.replace(["ERROR", "UNKNOWN"], np.nan, inplace=True)
```

```python
df.isnull().sum()
```

```python
numeric_cols = ["Quantity", "Price Per Unit", "Total Spent"]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
df["Transaction Date"] = pd.to_datetime(
    df["Transaction Date"],
    errors="coerce"
)
```

```python
for col in ["Quantity", "Price Per Unit"]:
    df[col].fillna(df[col].median())
```

```python
mask = df["Total Spent"].isna()

df.loc[mask, "Total Spent"] = (
    df.loc[mask, "Quantity"] *
    df.loc[mask, "Price Per Unit"]
)
```

```python
df["Item"].fillna(df["Item"].mode()[0])
```

```python
df["Payment Method"] = df["Payment Method"].ffill()
```

```python
df["Location"] = df["Location"].fillna("Unknown")
```

```python
df["Transaction Date"] = df["Transaction Date"].ffill()
```

```python
numeric_cols = ["Quantity", "Price Per Unit", "Total Spent"]

df[numeric_cols].describe()
```

```python
import matplotlib.pyplot as plt

for col in numeric_cols:
    plt.figure(figsize=(6,4))
    plt.boxplot(df[col])
    plt.title(f"Boxplot of {col}")
    plt.xlabel(col)
    plt.show()
```

```python
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]

    print(f"{col}: {len(outliers)} outliers")
```
